```python
def get_playlist_tracks(username, playlist_id):
    results = sp.user_playlist_tracks(username,playlist_id,market="GB")
    tracks = results['items']
    while results['next']:
        results = sp.next(results)
        tracks.extend(results['items'])
    return tracks
```

```python
def make_dataframe(playlist):
    df=pd.DataFrame()
    b=0
    for i in playlist:
        b+=1
        temp_df=pd.DataFrame(sp.audio_features(i["track"]["uri"])[0], index=[0])
        temp_df["main_artist"]=i["track"]["artists"][0]["name"]
        temp_df["artist_id"]=i["track"]["artists"][0]["id"]
        temp_df["artist_genre"]=[sp.artist(i["track"]["artists"][0]["id"])["genres"]]
        df=pd.concat([df, temp_df], ignore_index=True)
        if b%100==0:
            print(b)
    return df
```
